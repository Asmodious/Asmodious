```python
##lets move onto visualization
plt.figure(figsize=(6,6))
sns.pairplot(df)
```

```python
#from the above pair of graphs the store area and item available has relatively in a linear relation

sns.regplot(df['Store_Area'],df['Items_Available'] )
plt.title('Store_Area vs Items_Available', fontsize=20)
plt.xlabel('Store_Area')
plt.ylabel('Items_Available')
plt.show()
```

```python
#lets check the correlation of the dataframe
df.corr()
```

```python
#lets look through the heatmap
df_corr=df.corr(method='pearson')
plt.figure(figsize=(8,8))
sns.heatmap(df_corr,annot=True)
```
